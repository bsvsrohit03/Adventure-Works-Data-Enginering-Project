### NOTE: This is an alternative connection method using dbutils.
### This project uses spark.conf.set() instead.
### Included for reference only.

```python
# Mount bronze container
dbutils.fs.mount(
    source = "abfss://bronze@awdatastore.dfs.core.windows.net/",
    mount_point = "/mnt/bronze",
    extra_configs = {
        "fs.azure.account.auth.type": "OAuth",
        "fs.azure.account.oauth.provider.type": 
            "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
        "fs.azure.account.oauth2.client.id": "your_client_id_here",
        "fs.azure.account.oauth2.client.secret": "your_client_secret_here",
        "fs.azure.account.oauth2.client.endpoint": 
            "https://login.microsoftonline.com/your_tenant_id/oauth2/token"
    }
)
```
